```python
ab['Review Rating'] = ab.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
ab.isnull().sum()
```

```python
ab.columns = ab.columns.str.lower()
ab.columns = ab.columns.str.replace(' ' ,'_')
ab = ab.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
ab.columns
```

```python
labels = ['Young Adult', 'Adult','Middle-aged','Senior']
ab['age_group'] = pd.qcut(ab['age'],q=4,labels=labels)
```

```python
ab[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly'  : 7,
    'Monthly' : 30,
    'Quaterly'  : 90,
    'Bi-Weekly'  : 14,
    'Annualy' : 365,
    'Every 3 month' : 90
}

ab['purchase_frequency_days']  = ab['frequency_of_purchases'].map(frequency_mapping)
```

```python
ab[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
